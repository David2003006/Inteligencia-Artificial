```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df_titanic = pd.read_csv("Sets/train.csv")
df_test_titanic = pd.read_csv("Sets/test.csv")
id = df_test_titanic['PassengerId']
df_test_titanic
```

```python
mode = df_test_titanic['Embarked'].mode()
print(mode)
```

```python
#df_titanic.isnull().sum()

mean_age = df_test_titanic['Age'].mean()
df_test_titanic['Age'] = df_test_titanic['Age'].fillna(df_test_titanic['Age'].mean())
df_test_titanic['Cabin'] = df_test_titanic['Cabin'].fillna(df_test_titanic['Cabin'].mode())
df_test_titanic['Embarked'] = df_test_titanic['Embarked'].fillna('S')
df_test_titanic['Age'] = df_test_titanic['Age'].apply(lambda x: mean_age if x < 1 else x)
    
#print("Se llenaron los valores nulos correctamente.")
#df_titanic.isnull().sum()
```

```python

df_test_titanic['Family_Size'] = df_test_titanic['SibSp'] + df_test_titanic['Parch'] + 1
df_test_titanic['IsAlone'] = (df_test_titanic['Family_Size'] == 1).astype(int)
df_test_titanic['Title'] = df_test_titanic['Name'].str.extract(',\s*([^ ]+)\s', expand=False)
df_test_titanic['Fare_Per_Person'] = df_test_titanic['Fare'] / df_test_titanic['Family_Size']
df_test_titanic['Fare'] = pd.cut(df_test_titanic['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very_High'])
df_test_titanic['Embarked_Class'] = df_test_titanic['Embarked'].astype(str) + "_" + df_test_titanic['Cabin'].astype(str)
df_test_titanic['Age_Category'] = pd.cut(df_test_titanic['Age'], bins=[0, 12, 18, 35, 60, 120], labels=['Child', 'Teen', 'Adult', 'Middle_Aged', 'Senior'])
    
```

```python
columns = ['Age', 'Name', 'SibSp', 'Parch', 'Cabin', 'Embarked', 'Ticket', 'PassengerId']

df_test_titanic.drop(columns=columns, inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import  GridSearchCV
from sklearn.metrics import classification_report


x = df_titanic.drop(columns='Survived')
y= df_titanic['Survived']

rf = RandomForestClassifier(random_state=42)

param_grid = {
        'n_estimators': [50, 100, 200],           # Número de árboles
        'max_depth': [None, 10, 20, 30],          # Profundidad máxima del árbol
        'min_samples_split': [2, 5, 10],          # Mínimo de muestras requeridas para dividir un nodo
        'min_samples_leaf': [1, 2, 4],            # Mínimo de muestras en una hoja
        'max_features': ['log2', 'sqrt', None]
    }

grid_search= GridSearchCV(estimator=rf, param_grid= param_grid, 
                          scoring='accuracy', cv=5, n_jobs=1,verbose=2 )
grid_search.fit(x,y)

print("Mejores hiperparametros:", grid_search.best_estimator_)
print("Mejor presicion:", grid_search.best_score_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score

x = df_titanic.drop(columns='Survived')
y= df_titanic['Survived']

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = RandomForestClassifier(max_features=None, min_samples_leaf=2,
                       min_samples_split=10, n_estimators=200, random_state=42)

model.fit(X_train, Y_train)
y_predict = model.predict(X_test)

acurrancy = accuracy_score(Y_test, y_predict)
f1 = f1_score(Y_test, y_predict)

print(f"La presicion es de: {acurrancy}")
print(f"Mi F1 es de: {f1}")
print(y_predict)
```

```python
columns_train = model.feature_names_in_

df_test_titanic= df_test_titanic.reindex(columns= columns_train, fill_value=0)

y_predict = model.predict(df_test_titanic)

df_output= pd.DataFrame({
    'id':  id,
    'Survived': y_predict
})

df_output.to_csv("resultado.csv", index=False)
print("Archivo correctamente creado....")
```
