```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df_titanic = pd.read_csv("Sets/train.csv")
df_titanic = df_titanic[['Survived','Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
```

```python
df_titanic
```

```python
df_titanic= pd.get_dummies(df_titanic, columns=['Sex', 'Embarked'])
```

```python
df_titanic
```

```python
mean_age = df_titanic['Age'].mean()
df_titanic['Age'] =  df_titanic['Age'].fillna(mean_age)
df_titanic.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report

x = df_titanic.drop(columns='Survived')
y= df_titanic['Survived']

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)
y_predict = model.predict(X_test)

acurrancy = accuracy_score(Y_test, y_predict)
f1 = f1_score(Y_test, y_predict)

print(f"La presicion es de: {acurrancy}")
print(f"Mi F1 es de: {f1}")
```

```python
print("Reporte de presiciones", classification_report(Y_test, y_predict))
```

```python
def Survived_titanic(df):
    predecir = df.values
    y_predict = model.predict(predecir)
    return y_predict
```

```python
def dataset_comvertido(df):
    df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
    df= pd.get_dummies(df, columns=['Sex', 'Embarked'])
    return df
    
```

```python
df_test = pd.read_csv("Sets/test.csv")
id= df_test['PassengerId'].values
df_titanic_test = dataset_comvertido(df_test)
prediciones_supervivencia = Survived_titanic(df_titanic_test)

df_output = pd.DataFrame({
    'id': id,
    'Survived': prediciones_supervivencia
})

df_output.to_csv("Sets/predicion.csv", index=False)
print("Archivo correctamente creado")
```
